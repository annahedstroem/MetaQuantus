```python
"""This cell contains mixed functions to support the loading and processing of different datasets."""
from typing import Union
import torch
import numpy as np
import matplotlib.pyplot as plt

def denormalise(
        img: np.ndarray,
        mean: np.ndarray = np.array([0.485, 0.456, 0.406]), # [0.5092, 0.5171, 0.5500]
        std: np.ndarray = np.array([0.229, 0.224, 0.225]), # [0.2542, 0.2514, 0.2675]
        **kwargs
) -> np.ndarray:
    """De-normalise a torch image (using conventional ImageNet values)."""
    return (np.array(img) * std.reshape(-1, 1, 1)) + mean.reshape(-1, 1, 1)


def plot_inputs(x_batch: np.array,
                y_batch: np.array,
                nr_images: int = 5,
                img_size: int = 28,
                name_dataset: str = "",
                mapping: Union[None, dict] = None):
    fig, axes = plt.subplots(nrows=1, ncols=nr_images, figsize=(nr_images *3, int(nr_images * 2 /3)))

    for i in range(nr_images):
        if name_dataset == "ImageNet":
            y_name = str(mapping[y_batch[i]]).split(",")[0]
            axes[i].imshow((np.moveaxis(denormalise(x_batch[i]), 0, -1) * 255).astype(np.uint8), vmin=0.0, vmax=1.0, cmap="gray")
        else:
            y_name = str(y_batch[i].item())
            axes[i].imshow(x_batch[i].reshape(img_size, img_size), vmin=0.0, vmax=1.0, cmap="gray")
        axes[i].title.set_text(f"{name_dataset} - {y_name}")
        axes[i].axis("off")
    plt.show()


def evaluate_model(model, data, device):
    """Evaluate torch model."""
    model.eval()
    logits = torch.Tensor().to(device)
    targets = torch.LongTensor().to(device)

    with torch.no_grad():
        for b, (images, labels) in enumerate(data):
            images, labels = images.to(device), labels.to(device)
            logits = torch.cat([logits, model(images)])
            targets = torch.cat([targets, labels])

    return torch.nn.functional.softmax(logits, dim=1), targets


def train_model(model, train_loader, test_loader, criterion, optimizer, epochs, path_model, device):
    """Train model."""
    for epoch in range(epochs):
        model.train()

        for b, (images, labels) in enumerate(train_loader):

            images, labels = images.to(device), labels.to(device)
            logits = model(images)

            loss = criterion(logits, labels)
            model.zero_grad()
            loss.backward()
            optimizer.step()

        # Evaluate model!
        predictions, labels = evaluate_model(model, test_loader, device)
        test_acc = np.mean(np.argmax(predictions.cpu().detach().numpy(), axis=1) == labels.detach().cpu().numpy())
        print(f"Epoch {epoch +1}/{epochs} - loss {loss.item():.2f} - test accuracy: {(100 * test_acc):.2f}%")

    # Save model.
    torch.save(model.state_dict(), path_model)

    return model
```

```python
"""This cell contains functions to create segmentation masks for MNIST and fMNIST datasets."""
import numpy as np
import cv2

def create_bbox_mnist(x_batch: np.array,
                      eps: float = 0.5,
                      img_size: int = 28,
                      ) -> np.array:
    """

    Returns
    -------
    Mask of a bounding box around an image.
    """
    s_batch = []
    for x in x_batch:

        # Apply the threshold:
        binaryImage = np.zeros_like(x[0], dtype=np.uint8)
        binaryImage[x[0] > np.finfo(float).eps] = 1
        b1 = np.min(np.argwhere(np.sum(binaryImage, axis=0) > np.finfo(float).eps))
        b2 = np.max(np.argwhere(np.sum(binaryImage, axis=0) > np.finfo(float).eps))
        b3 = np.min(np.argwhere(np.sum(binaryImage, axis=1) > np.finfo(float).eps))
        b4 = np.max(np.argwhere(np.sum(binaryImage, axis=1) > np.finfo(float).eps))

        output = np.zeros(x[0].shape)
        cv2.rectangle(output, (b1, b3), (b2, b4), 1, -1)
        s_batch.append(output.reshape(x.shape).astype(int))
    return np.array(s_batch)


def create_mask_mnist(x_batch: np.array,
                      eps: float = 0.5,
                      img_size: int = 28,
                      ) -> np.array:
    """

    Returns
    -------
    Mask 1=non-zero pixels, 0=zero pixels.
    """
    s_batch = []
    for x in x_batch:
        # Apply the threshold:
        binaryImage = np.zeros_like(x[0], dtype=np.uint8)
        binaryImage[x[0] > np.finfo(float).eps] = 1
        s_batch.append(binaryImage.reshape(x.shape).astype(int))

    return np.array(s_batch)


def create_dilated_mask_mnist(x_batch: np.array,
                      kernel_size: int = 3,
                      ) -> np.array:
    """

    Returns
    -------
    Digit mask plus a margin in all directions.
    """
    s_batch = []
    kernel = np.ones((kernel_size, kernel_size), np.uint8)
    for x in x_batch:
        # Apply the threshold:
        binaryImage = np.zeros_like(x[0], dtype=np.uint8)
        binaryImage[x[0] > np.finfo(float).eps] = 1
        img_dilation = cv2.dilate(binaryImage, kernel, iterations=1)
        s_batch.append(img_dilation.reshape(x.shape).astype(int))

    return np.array(s_batch)
```

```python
"""This cell contains functions to create the data for cMNIST dataset."""

import torch
import random
import scipy.ndimage
from typing import Tuple, Dict
import math
import numpy as np
import pickle


class TransformCMNIST(torch.utils.data.Dataset):
    """Class to transform cMNIST dataset."""

    def __init__(self, path_dict, transform=None, noise=False, mean=0.0, std=0.0, nr_channels=3, img_size=32):

        self.path_dict = path_dict
        self.transform = transform
        self.noise = noise
        self.mean = mean
        self.std = std

        with open(self.path_dict, 'rb') as handle:
            data = pickle.load(handle)

        samples = len(data)
        self.Y = torch.zeros(samples)
        self.X = torch.zeros([samples, nr_channels, img_size, img_size])
        self.s_box = torch.zeros([samples, nr_channels, img_size, img_size])
        self.s_digit = torch.zeros([samples, nr_channels, img_size, img_size])
        self.s_area = torch.zeros([samples, nr_channels, img_size, img_size])

        for index in range(len(data)):
            self.X[index] = torch.tensor(data[index]['x'])
            self.Y[index] = data[index]['y']

            self.s_box[index] = torch.tensor(data[index]['s_box'])
            self.s_digit[index] = torch.tensor(data[index]['s_digit'])
            self.s_area[index] = torch.tensor(data[index]['s_area'])

    def __getitem__(self, idx):
        if self.transform is not None:
            X = self.transform(self.X[idx])
        else:
            X = self.X[idx]

        if self.noise:
            return X + torch.randn(X.size()) * self.std + self.mean, self.Y[idx].long()
        else:
            return X, self.Y[idx].long()

    def __len__(self):
        return len(self.X)

    def get_digit(self, idx):
        return self.s_digit[idx]

    def get_box(self, idx):
        return self.s_box[idx]

    def get_area(self, idx):
        return self.s_area[idx]


def _mapping_uniform_sampling_cifar_backgrounds(cifar: torch.utils.data.Dataset,
                                               mnist: torch.utils.data.Dataset) -> Dict:
    """Get the indices per class for CIFAR and MNIST."""
    indices = {}
    indices_mnist = {}
    for c in range(10):
        l = []
        for i_c, (_, y) in enumerate(cifar):
            if y == c:
                l.append(i_c)
        indices[c] = l
        l = []
        for i_m, (_, y) in enumerate(mnist):
            if y == c:
                l.append(i_m)
        indices_mnist[c] = l

    # Calculate how many CIFAR background needed per MNIST digit (slightly imbalanced classes).
    nr_cifars = [math.ceil(len(v) / 10) for k, v in indices_mnist.items()]

    # Re-collect random CIFAR indices, enough per class.
    random_samples = {}
    for nx, nr in enumerate(nr_cifars):
        l = []
        for ix, (k, v) in enumerate(indices.items()):
            samples = random.sample(v, nr)
            l.append(samples)
        l = [i for sl in l for i in sl]
        np.random.shuffle(l)
        random_samples[nx] = l

    return random_samples


def _get_neighbouring_pixels(digit: torch.Tensor, img_size: int = 32, pixels_add: int = 3, **kwargs) -> torch.Tensor:
    """Flip digit area, 3 pixels from digit pixel in x, y direction."""

    area = torch.zeros_like(digit)
    non_zeros = digit.nonzero()

    for i in non_zeros[:int(len(non_zeros) / pixels_add)]:
        for c, y, x in [i]:
            y_index_up = int(y) - pixels_add
            y_index_down = int(y) + pixels_add

            x_index_left = int(x) - pixels_add
            x_index_right = int(x) + pixels_add

            if y_index_up < 0:
                y_index_up = 0
            if y_index_down >= img_size:
                y_index_down = img_size
            if x_index_left < 0:
                x_index_left = 0
            if x_index_right >= img_size:
                x_index_right = img_size

            # Add pixels in all directions.
            area[c, y, x_index_left: x] = 1.0
            area[c, y, x: x_index_right] = 1.0
            area[c, y_index_up: y, x] = 1.0
            area[c, y: y_index_down, x] = 1.0
            area[c, y_index_up: y, x_index_left: x] = 1.0
            area[c, y_index_up: y, x: x_index_right] = 1.0
            area[c, y: y_index_down, x_index_left: x] = 1.0
            area[c, y: y_index_down, x: x_index_right] = 1.0

    area[area != 0] = 1
    area[area == 0] = 0

    return area


def save_cmnist_dict_disk(mnist: torch.utils.data.Dataset,
                          cifar: torch.utils.data.Dataset,
                          img_size: int = 32,
                          digit_size: int = 16,
                          rotate: Tuple[int, int] = [15, 15],
                          noise: bool = True,
                          path_dict: str = "/content/drive/MyDrive/Projects/assets/datasets/cmnist_dict_dataset.pickle",
                          **kwargs):
    data = {}
    class_samples = _mapping_uniform_sampling_cifar_backgrounds(cifar, mnist)

    for index, (x, y) in enumerate(mnist):

        # Get digit and apply rotation.
        digit = x.transpose(0, 1).transpose(1, 2).repeat([1, 1, 3])
        digit = torch.tensor(scipy.ndimage.rotate(digit, np.random.uniform(rotate[0], rotate[1]), reshape=False))
        digit = digit.transpose(2, 1).transpose(1, 0)

        # Get a random CIFAR index per class (to be uniformly distributed).
        class_indices = class_samples[y]
        cifar_index = random.choice(class_indices)
        class_indices.remove(cifar_index)
        class_samples[y] = class_indices

        # Get image and apply customisation.
        background = cifar[cifar_index][0]
        coords = np.random.randint(0, img_size - digit_size, 2)
        eps = 0.05

        image = background.clone().detach()
        image[:, coords[0]: coords[0] + digit_size, coords[1]: coords[1] + digit_size][digit > eps] = \
            image[:, coords[0]: coords[0] + digit_size, coords[1]: coords[1] + digit_size][digit > eps] * 0.1 + digit[
                digit > eps] * 0.9

        # Get box and segmentation indices.
        s_box = torch.zeros_like(image)
        s_box[:, coords[0]: coords[0] + digit_size, coords[1]: coords[1] + digit_size] = 1.0
        s_digit = torch.zeros_like(image)
        s_digit[:, coords[0]: coords[0] + digit_size, coords[1]: coords[1] + digit_size][digit > eps] = 1.0
        s_area = _get_neighbouring_pixels(digit=s_digit, img_size=img_size,
                                         pixel_add=kwargs.get("pixel_add", 3), **kwargs)

        # Add noise and normalise.
        if noise:
            image = image + torch.randn(image.shape) / random.randint(10, 30)  # 15
            image[image < 0] = 0.
            image[image > 1] = 1.

        # Append to data dictionary.
        data[index] = {"id": index,
                       "x": image.numpy(),
                       "y": y,
                       "s_box": s_box.numpy(),
                       "s_digit": s_digit.numpy(),
                       "s_area": s_area.numpy()}

    return data

```

```python
"""This cell contains functions to generate the data for ImageNet dataset. 
For this you have to have ImageNet data and its associated annotation files downloaded locally.

For more information, please read more here: https://www.image-net.org/download.php.
"""


def get_imagenet_int_labels(path: str = "/content/drive/MyDrive/Projects/quantus/tutorials/assets"):
    """Make a int-string label mapping for Imagenet classes."""
    mapping_int = {}
    with open(f'{path}imagenet_images/imagenet_labels.txt', 'r') as f:
        for ix, line in enumerate(f):
            if ix not in [0, 1001]:
                line = line.split('b" ')[0]
                key = line.split(':')[0]
                value = line.split(": '")[1].split("',\n")[0]
                mapping_int[int(key)] = str(value)

    return mapping_int


def get_imagenet_str_labels(folder_names: List[str],
                            path: str = "drive/MyDrive/Projects/quantus/tutorials/assets"):
    mapping_str = {}
    with open(path + "/imagenet_images/imagenet1000_clsid_to_labels.txt", "r") as f:
        lines = f.readlines()
        for l in lines:
            line_strings = l.split(":")
            c = str(line_strings[0])
            if c in folder_names:
                mapping_str[c] = str(line_strings[1][2:-3])

    return mapping_str


def get_labels(mapping_str: Dict[str, str], mapping_int: Dict[str, str]):
    labels = []
    for label_int, label_str_1 in mapping_int.items():
        for label_str, label_str_2 in mapping_str.items():
            if label_str_1 == label_str_2:
                labels.append(label_int)
    return labels

def load_sample_edit(path: str) -> dict:
    """Read data as image and path. """
    return {"input": PIL.Image.open(path).convert("RGB"), "path": path}


def load_binary_mask_edit(filename: str, path_data: str):
    """Load the binary mask for the given path of the data in the correct format. """
    binary_mask = {}

    filename = os.path.splitext(filename)[0]

    # Get label and file name.
    label = filename.split("/")[-2]
    fname = filename.split("_")[-1]

    # Parse annotations.
    full_fname_annotation = os.path.join(path_data, "Annotations/{}/{}_{}.xml".format(label, label, fname))
    full_fname_image = os.path.join(path_data, f"bboxes_images/{label}/{label}_{fname}.JPEG")

    if os.path.isfile(full_fname_image) and os.path.isfile(full_fname_annotation):
        #print(f"{full_fname_annotation} AND {full_fname_image} its a MATCH!!!!")

        tree = ElementTree.parse(full_fname_annotation)
        xml_data = tree.getroot()
        xmlstr = ElementTree.tostring(xml_data, encoding="utf-8", method="xml")
        annotation = dict(xmltodict.parse(xmlstr))['annotation']

        width = int(annotation["size"]["width"])
        height = int(annotation["size"]["height"])

        # Iterate objects.
        objects = annotation["object"]

        if type(objects) != list:
            mask = np.zeros((height, width), dtype=int)
            mask[int(objects['bndbox']['ymin']):int(objects['bndbox']['ymax']),
            int(objects['bndbox']['xmin']):int(objects['bndbox']['xmax'])] = 1
            binary_mask[objects['name']] = mask

        else:
            for object in annotation['object']:
                if type(object) == collections.OrderedDict:
                    if object['name'] in binary_mask.keys():
                        mask = binary_mask[object['name']]
                    else:
                        mask = np.zeros((height, width), dtype=np.uint8)

                    mask[int(object['bndbox']['ymin']):int(object['bndbox']['ymax']),
                    int(object['bndbox']['xmin']):int(object['bndbox']['xmax'])] = 1

                    binary_mask[object['name']] = mask

        # Preprocess binary masks to fit shape of image data.
        for key in binary_mask.keys():
            binary_mask[key] = cv2.resize(binary_mask[key],
                                          (224, 224),
                                          interpolation=cv2.INTER_NEAREST).astype(np.int)[:, :, np.newaxis]

        return binary_mask, label
    else:
        return None

class CustomTransformEdit(torch.nn.Module):
    """Custom transformation to handle image processing and binary mask processing simultaneously. """

    # Collect garbage.
    gc.collect()
    torch.cuda.empty_cache()
    
    image_transform = transforms.Compose([transforms.Resize((224, 224)),
                                          transforms.ToTensor(),
                                          transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])])

    def __init__(self, path):
        super().__init__()
        self.path = path

    def __call__(self, img, **kwargs):
        return {
            "image": self.image_transform(img["input"]),
            "mask_label": (load_binary_mask_edit(img["path"], self.path)),
            }

def CustomCollateEdit(batch):
    """Collate function to pack the image, mask and label data accordingly. """
    
    MAX_SIZE = 128
    inputs = []
    targets = []
    masks = []
    
    LABELS = [13, 96, 126, 292, 301, 308, 347, 387, 394, 405, 417, 426, 449, 470, 490, 516, 538, 615, 856]
    
    #print(len(batch))
    for b in batch:
        # print(type(b[0]["mask"]), type(b), np.shape(b), b[1])#, LABELS[b[1]])
        if b[0]["mask_label"] is not None:
            mask = b[0]["mask_label"][0]#, b[0]["mask_label"][1]
            if len(mask) > 0:
                image = b[0]["image"]
                inputs.append(image)
                masks.append(mask)
                try:
                    #print("targets=ok -- label", label, b[1], LABELS[b[1]])
                    targets.append(LABELS[b[1]])
                except:
                    targets.append(None)

    # Collect garbage.
    gc.collect()
    torch.cuda.empty_cache()

    # Rearrange masks, inputs and targets.
    masks = torch.Tensor(np.array([m[list(m.keys())[0]][:, :, 0] for m in masks]))[:MAX_SIZE]
    inputs = torch.stack(inputs, 0)[:MAX_SIZE]
    targets = torch.tensor(targets)[:MAX_SIZE]

    # Collect garbage.
    gc.collect()
    torch.cuda.empty_cache()

    return inputs, masks, targets
```

```python
# Save the data, batch by batch for performance efficiency.
for batch_i in range(8):

    if batch_i < 2:
        continue

    path_imagenet_assets = path + f"test_sets/imagenet_test_set_{batch_i}_128.npy"

    print(f"Processing batch {int(batch_i+1)} ... saving --> {path_imagenet_assets} ")

    # Settings data.
    dataset = torchvision.datasets.DatasetFolder(os.path.join("drive/MyDrive/Projects/imagenet_images/", "bboxes_images/"),
                                                loader=load_sample_edit,
                                                extensions=(".jpeg", "png"),
                                                transform=CustomTransformEdit(path="drive/MyDrive/Projects/imagenet_images/")
                                                )
    test_loader = torch.utils.data.DataLoader(dataset, shuffle=True, batch_size=batch_size_test, collate_fn=CustomCollateEdit)

    # Load a batch of inputs, segmentation masks and outputs to use for evaluation.
    x_batch, s_batch, y_batch = iter(test_loader).next()
    x_batch, s_batch, y_batch = x_batch.cpu().numpy(), s_batch.cpu().numpy(), y_batch.cpu().numpy()
    plot_inputs(x_batch=x_batch, y_batch=y_batch, nr_images=5, img_size=224, name_dataset="ImageNet", 
                mapping=get_imagenet_int_labels("drive/MyDrive/Projects/"))

    # Collect garbage.
    gc.collect()
    torch.cuda.empty_cache()

    print(f"Length of batch {len(x_batch)}")

    # Save model, x_batch, y_batch, s_batch to one dictionary.
    assets = {'x_batch' : x_batch,
             'y_batch': y_batch,
             's_batch': s_batch}

    np.save(path_imagenet_assets,  assets) 

```
